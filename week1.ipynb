### Week 1 - Monthly Dataset Aggregation

```python
import pandas as pd
import glob #grabs multiple files at once automatically
import os #helps with files paths
```

```python
folder_path = "data/raw"

sold_files = glob.glob(os.path.join(folder_path, "CRMLSSold*.csv"))
listing_files = glob.glob(os.path.join(folder_path, ("CRMLSListing*.csv")))

#the * helps match anything, which specifically helps with locating the difference in listing and sold files. 
#glob.glob helps find the files regarding the pattern it needs to follow. 

print(len(sold_files))
print(len(listing_files))
```

```python
sold_list = []

for file in sold_files:
    new_df = pd.read_csv(file)
    sold_list.append(new_df)

sold = pd.concat(sold_list, ignore_index = True)
print(f"Sold before filtering (rows,columns): {sold.shape}")


listing_list = []

for file in listing_files: 
    new_df = pd.read_csv(file)
    listing_list.append(new_df)

listing = pd.concat(listing_list, ignore_index= True)
print(f"Listings before filtering (rows,columns): {listing.shape}")


# .concat() helps create one big dataframe
# ignore_index = True helps with duplicate rows with the same index position, it'll restart the rows for the new df
```

```python
#Intern Questions: What is the Residential vs. other property type share?

print(f"Sold: {sold['PropertyType'].value_counts(normalize=True)*100}")
print(f"Listing: {listing['PropertyType'].value_counts(normalize=True)*100}")
```

```python
#filtering to Residential

sold = sold[sold['PropertyType'] == 'Residential']
listing = listing[listing['PropertyType'] == 'Residential']

print(f"Sold after filtering (rows, columns): {sold.shape}")
print(f"Listing after filtering (rows, columns): {listing.shape}")
```

```python
#saving new combined datasets to new CSVs

sold.to_csv("data/sold_combined.csv", index = False)
listing.to_csv("data/listing_combined.csv", index = False)
```
